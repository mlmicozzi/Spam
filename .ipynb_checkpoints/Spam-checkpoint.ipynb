```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# Scikit-learn nos ofrece una variedad ampliada de modelos Naive Bayes, para este problema usamos MultinomialNB que es pensado para este tipo de problemas
from sklearn.naive_bayes import MultinomialNB   

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
dataset = pd.read_csv("spambase.csv") # cargando los datos desde un CSV
dataset.head(10)
```

```python
dataset.info()
```

```python
dataset["spam"].value_counts() 
```

```python
dataset["spam"].value_counts(normalize=True) * 100
```

#### Punto 1:

```python
column_sum = dataset.groupby(by="spam", as_index=False).sum()
column_sum
```

```python
melt = pd.melt(column_sum, id_vars='spam', var_name='word', value_name='frequency')

spam_words = melt[melt['spam'] == 1].sort_values(by='frequency', ascending=False).head(10)
no_spam_words = melt[melt['spam'] == 0].sort_values(by='frequency', ascending=False).head(10)
```

```python
print("Las 10 palabras más encontradas en correo SPAM son:")
spam_words[["word" , "frequency"]]
```

```python
print("Las 10 palabras más encontradas en correo No SPAM son:")
no_spam_words[["word" , "frequency"]]
```

```python
# Palabras en común del top 10:

# Obtener valores únicos en la columna 'word'
words_spam = set(spam_words['word'])
words_no_spam = set(no_spam_words['word'])

# Encontrar valores comunes
valores_comunes = words_spam.intersection(words_no_spam)
valores_comunes
```

```python
# Obtenemos los atributos y target
X = (dataset.drop(columns="spam") * 100).astype(int)
#X = dataset2.drop(columns="spam")
y = dataset["spam"]
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size = 0.3 , random_state=42)
```

```python
# Creamos el modelo de Naive Bayes Multinomial:
model_nb = MultinomialNB()
```

```python
# Entrenamos el modelo:
model_nb.fit(X_train, y_train)
```

```python
# Normalizamos los datos:
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Lo transformamos en DataFrames
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X.columns)
```

```python
# Creamos el modelo de regresión logística
model_lr = LogisticRegression()

# Entrenamos el modelo con los datos de entrenamiento
model_lr.fit(X_train_scaled, y_train)
```

```python
y_pred_nb = model_nb.predict(X_test)
cm_nb = confusion_matrix(y_test, y_pred_nb)
print("Matriz de confusión - Bayes ingenuo:")
print(cm_nb)
```

```python
# Matriz de confusión para Bayes ingenuo:
disp_nb = ConfusionMatrixDisplay(confusion_matrix=cm_nb, display_labels=['No SPAM', 'SPAM'])
disp_nb.plot(cmap='Blues')
plt.title("Matriz de Confusión - Bayes ingenuo")
plt.show()
```

```python
y_pred_lr = model_lr.predict(X_test_scaled)
cm_lr = confusion_matrix(y_test, y_pred_lr)
print("Matriz de confusión - Logistic Regression:")
print(cm_lr)
```

```python
# Matriz confusión para el modelo de Regresión Logística:
disp_lr = ConfusionMatrixDisplay(confusion_matrix=cm_lr, display_labels=['No SPAM', 'SPAM'])
disp_lr.plot(cmap='Reds')
plt.title("Matriz de Confusión - Regresión Logística")
plt.show()
```

#### Conclusión:
##### En este ejemplo, lo que interesa es minimizar el error de Tipo 1, dado que se considera más grave etiquetar un correo legítimo como spam. En cambio, no es tan grave que un correo spam lo etiquete como legítimo, dado que el usuario, puede hacer un chequeo final.
##### El modelo Naive Bayes comete más error de Tipo 1, mientras que el modelo de Regresión Logística comete más error de Tipo 2.

```python
accuracy_nb = accuracy_score(y_test, y_pred_nb)
precision_nb = precision_score(y_test, y_pred_nb)
recall_nb = recall_score(y_test, y_pred_nb)
f1_score_nb = f1_score(y_test, y_pred_nb)
print(f"Accuracy - Bayes ingenuo: {accuracy_nb:.4f}")
print(f"Precisión - Bayes ingenuo: {precision_nb:.4f}")
print(f"Recuperación - Bayes ingenuo: {recall_nb:.4f}")
print(f"f1 Score - Bayes ingenuo: {f1_score_nb:.4f}")
```

```python
accuracy_lr = accuracy_score(y_test, y_pred_lr)
precision_lr = precision_score(y_test, y_pred_lr)
recall_lr = recall_score(y_test, y_pred_lr)
f1_score_lr = f1_score(y_test, y_pred_lr)
print(f"Accuracy - Regresión Logística: {accuracy_lr:.4f}")
print(f"Precisión - Regresión Logística: {precision_lr:.4f}")
print(f"Recuperación - Regresión Logística: {recall_lr:.4f}")
print(f"f1 Score - Regresión Logística: {f1_score_lr:.4f}")
```

#### Conclusión:
##### Para la métrica de Presición es mejor el modelo de Regresión Logística y para la Recuperación, se obtienen mejores resultados con el modelo Naive Bayes.
##### La métrica de Precisión se relaciona más con el error de Tipo 1, mientras que la métrica de Recuperación, se relaciona más con el error de Tipo 2.
##### Como en este caso se prioriza minimizar el error de Tipo 1, la métrica que más nos interesa es Precisión. 

```python
from sklearn.metrics import roc_curve, auc

y_prob_nb = model_nb.predict_proba(X_test_scaled)
y_prob_lr = model_lr.predict_proba(X_test_scaled)

# Curva ROC para Naive Bayes:
fpr_nb, tpr_nb, _ = roc_curve(y_test, y_prob_nb[:, 1])

# Calculamos el AUC:
roc_auc_nb = auc(fpr_nb, tpr_nb)
print('AUC para Naive Bayes = %0.2f' % roc_auc_nb)
```

```python
# Curva ROC para Regresión Logística:
fpr_lr, tpr_lr, _ = roc_curve(y_test, y_prob_lr[:, 1])

# Calcular el AUC
roc_auc_lr = auc(fpr_lr, tpr_lr)
print('AUC para Regresión Logística = %0.2f' % roc_auc_lr)
```

```python
# Graficar la curva ROC
plt.figure()
plt.plot(fpr_lr, tpr_lr, color='red', lw=2, label='Regresión Logística (AUC = %0.2f)' % roc_auc_lr)
plt.plot(fpr_nb, tpr_nb, color='blue', lw=2, label='Naive Bayes (AUC = %0.2f)' % roc_auc_nb)
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Curva ROC') 
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
